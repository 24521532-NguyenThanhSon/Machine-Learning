```python
from __future__ import print_function
import numpy as np
from sklearn import neighbors,datasets
from sklearn.model_selection import train_test_split # for splitting data
from sklearn.metrics import accuracy_score # for evaluating results
np.random.seed(7)
```

```python
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
print('Labels:', np.unique(iris_y))
# split train and test
X_train, X_test, y_train, y_test = train_test_split(iris_X, iris_y, test_size = 130)
print('Train Test: ', X_train.shape[0], '  Testsize: ',X_test.shape[0])
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=1, p = 2)
model.fit(X_train, y_train)
y_pred =  model.predict(X_test)
print(100 *accuracy_score(y_test,y_pred))
# KNeighbors, K-means_cluster
```

```python
import numpy as np
import matplotlib.pyplot as plt

# S im d liu mi cm
N = 500

# Ta  trung bnh ca cc cm
means = np.array([[2, 2], [8, 3], [3, 6]])

# Ma trn hip phng sai (n v - khng c tng quan gia cc bin)
cov = np.array([[1, 0], [0, 1]])

# Sinh d liu cho 3 cm
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
X2 = np.random.multivariate_normal(means[2], cov, N)

# V biu 
plt.scatter(X0[:, 0], X0[:, 1], color='red', alpha=0.5, label='Cluster 1')
plt.scatter(X1[:, 0], X1[:, 1], color='blue', alpha=0.5, label='Cluster 2')
plt.scatter(X2[:, 0], X2[:, 1], color='green', alpha=0.5, label='Cluster 3')

# Hin th tm cm
plt.scatter(means[:, 0], means[:, 1], color='black', marker='x', s=100, label='Centroids')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')

plt.title('Scatter plot of generated clusters')
plt.show()
```
