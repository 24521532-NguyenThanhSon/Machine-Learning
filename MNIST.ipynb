```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Đọc dữ liệu
X = pd.read_csv('mnist_test.csv')
# print(X['label'])
# Chọn ngẫu nhiên 10,000 mẫu
# X = mnist_data.sample(n=N, random_state=42)
# Loại bỏ cột nhãn (giả sử cột nhãn tên là 'label')
X_features = X.drop(columns=['label'], errors='ignore')
# print(X_features)
K = 10
kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(X_features)
# image_data = X_features.iloc[0].values
# image_data = image_data[:784]
# image_data = image_data.reshape(28,28)
# plt.imshow(image_data, cmap='gray')
# Dự đoán nhãn cụm cho các điểm dữ liệu
pred_label = kmeans.predict(X_features)
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
img = mpimg.imread('girl3.jpg')
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
X = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
print(X)
```
