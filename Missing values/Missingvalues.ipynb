```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
data = pd.read_csv('melb_data.csv')
y = data.Price
melb_predictors = data.drop(['Price'], axis = 1)
X = melb_predictors.select_dtypes(exclude = ['object'])
display(X)
display(melb_predictors)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size = 0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics  import mean_absolute_error
# Function for comparing diffent approaches 
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
#   Approach 1  Drop columns with missing values
cols_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()] 
# Drop columns in train and validation data
reduced_X_train = X_train.drop(cols_with_missing, axis = 1)
reduced_X_valid = X_valid.drop(cols_with_missing, axis = 1)
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))




```

```python
from sklearn.impute import SimpleImputer
# Imputation
my_imputer = SimpleImputer()
impute_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
impute_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
# impute drop columns
impute_X_train.columns = X_train.columns
impute_X_valid.columns = X_valid.columns
print(score_dataset(impute_X_train, impute_X_valid, y_train, y_valid))
```

```python
# Make copy to avoid changing original data (when imputing)
X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()

# Make new columns indicating what will be imputed
for col in cols_with_missing:
    X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()
    X_valid_plus[col + '_was_missing'] = X_valid_plus[col].isnull()
display(X_train_plus)
# Imputation
my_imputer = SimpleImputer()
imputed_X_train_plus = pd.DataFrame(my_imputer.fit_transform(X_train_plus))
imputed_X_valid_plus = pd.DataFrame(my_imputer.transform(X_valid_plus))

# Imputation removed column names; put them back
imputed_X_train_plus.columns = X_train_plus.columns
imputed_X_valid_plus.columns = X_valid_plus.columns

print("MAE from Approach 3 (An Extension to Imputation):")
print(score_dataset(imputed_X_train_plus, imputed_X_valid_plus, y_train, y_valid))
```
