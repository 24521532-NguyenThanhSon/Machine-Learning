```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Feature values (reshape for sklearn)
y = np.array([2, 4, 5, 4, 5])  # Actual target values

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Scatter plot of actual data
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the predicted points
plt.scatter(X, y_pred, color='red', label='Predicted Data', marker='x')

# Add labels and legend
plt.xlabel("X-axis (Feature)")
plt.ylabel("Y-axis (Target)")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.show()
```
